```python
#We conduct visualizations of the Survived variable with respect to other variables.
#Visualizations of the count of Survived with respect to Pclass.
# Converting 'Survived' and 'Pclass' to categorical
df['Survived'] = df['Survived'].astype(str)
df['Pclass'] = df['Pclass'].astype(str)

# Create the count plot
sns.countplot(x='Survived', hue='Pclass', data=df)

# Adding labels and title
plt.xlabel('Survived')
plt.ylabel('Count')
plt.title('Survival Count by Pclass')

# Displaying the plot
plt.show()
```

```python
#We then define the distribution of the sex(gender distribution.)
df["Sex"]
```

```python
#We then plot the sex distribution with respect to survival for a better understanding.
sns.countplot(data=df, x='Sex', hue='Survived')
plt.show()
```

```python
#Conducting the average of the gender survival rate using mean(Central Tendency).
# Convert 'Survived' to numeric (0 for not survived, 1 for survived)
df['Survived'] = df['Survived'].astype(int)

# Calculate the average survival rate by gender
survival_rate = df.groupby('Sex')['Survived'].mean()
print(survival_rate)
```

```python
#We need to convert the Sex variables from categorical to numerical values before building the model.
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()

df['Sex']= labelencoder.fit_transform(df['Sex'])
#We then visualize the first set of variables
df.head()
```

```python
#We can see that the age column is the only variable with missing values.
#I chose to drop the column rather than fill the missing values.
df=df.drop(['Age'], axis=1)
```

```python
df_final = df
df_final.head(20)
```

```python
#The age column is not available now.
#We fit the model 
#Model Fitting- LogisticRegression.
X= df[['Pclass', 'Sex']]
Y=df['Survived']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.20, random_state = 0)
```

```python
from sklearn.linear_model import LogisticRegression

log = LogisticRegression(random_state = 0)
log.fit(X_train, Y_train)
```

```python
pred = print(log.predict(X_test))
```
